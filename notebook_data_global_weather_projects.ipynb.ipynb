# Project Data Analysis & Visualization Menggunakan Global Weather Dataset

## Columns descriptions:

- country: Country of the weather data
- location_name: Name of the location (city)
- latitude: Latitude coordinate of the location
- longitude: Longitude coordinate of the location
- timezone: Timezone of the location
- last_updated_epoch: Unix timestamp of the last data update
- last_updated: Local time of the last data update
- temperature_celsius: Temperature in degrees Celsius
- temperature_fahrenheit: Temperature in degrees Fahrenheit
- condition_text: Weather condition description
- wind_mph: Wind speed in miles per hour
- wind_kph: Wind speed in kilometers per hour
- wind_degree: Wind direction in degrees
- wind_direction: Wind direction as a 16-point compass
- pressure_mb: Pressure in millibars
- pressure_in: Pressure in inches
- precip_mm: Precipitation amount in millimeters
- precip_in: Precipitation amount in inches
- humidity: Humidity as a percentage
- cloud: Cloud cover as a percentage
- feels_like_celsius: Feels-like temperature in Celsius
- feels_like_fahrenheit: Feels-like temperature in Fahrenheit
- visibility_km: Visibility in kilometers
- visibility_miles: Visibility in miles
- uv_index: UV Index
- gust_mph: Wind gust in miles per hour
- gust_kph: Wind gust in kilometers per hour
- air_quality_Carbon_Monoxide: Air quality measurement: Carbon Monoxide
- air_quality_Ozone: Air quality measurement: Ozone
- air_quality_Nitrogen_dioxide: Air quality measurement: Nitrogen Dioxide
- air_quality_Sulphur_dioxide: Air quality measurement: Sulphur Dioxide
- air_quality_PM2.5: Air quality measurement: PM2.5
- air_quality_PM10: Air quality measurement: PM10
- air_quality_us-epa-index: Air quality measurement: US EPA Index
- air_quality_gb-defra-index: Air quality measurement: GB DEFRA Index
- sunrise: Local time of sunrise
- sunset: Local time of sunset
- moonrise: Local time of moonrise
- moonset: Local time of moonset
- moon_phase: Current moon phase
- moon_illumination: Moon illumination percentage

```python
import pandas as pd

df = pd.read_csv("Global_Weather.csv")

df_clean = df.copy()

import pycountry_convert as pc
import pycountry


from sklearn.ensemble import IsolationForest

# ====================================
# LOAD DATASET
# ====================================

df = pd.read_csv("Global_Weather.csv")

print("Dataset berhasil dibuka!")
print(df.shape)

# ====================================
# ADD CONTINENT COLUMN
# ====================================

continent_map = {

    # ASIA
    "Indonesia": "Asia",
    "Japan": "Asia",
    "China": "Asia",
    "India": "Asia",
    "South Korea": "Asia",
    "Thailand": "Asia",
    "Malaysia": "Asia",
    "Singapore": "Asia",
    "Pakistan": "Asia",
    "Afghanistan": "Asia",

    # EUROPE
    "France": "Europe",
    "Germany": "Europe",
    "Italy": "Europe",
    "Spain": "Europe",
    "United Kingdom": "Europe",
    "Netherlands": "Europe",
    "Belgium": "Europe",
    "Switzerland": "Europe",

    # AFRICA
    "Nigeria": "Africa",
    "Egypt": "Africa",
    "South Africa": "Africa",
    "Kenya": "Africa",
    "Morocco": "Africa",

    # NORTH AMERICA
    "United States of America": "North America",
    "Canada": "North America",
    "Mexico": "North America",

    # SOUTH AMERICA
    "Brazil": "South America",
    "Argentina": "South America",
    "Chile": "South America",
    "Colombia": "South America",

    # OCEANIA
    "Australia": "Oceania",
    "New Zealand": "Oceania"
}

df["continent"] = df["country"].map(continent_map)

# unknown continent
df["continent"] = df["continent"].fillna("Unknown")

print("\nContinent column berhasil dibuat!")
print(df["continent"].value_counts())

# ====================================
# KEEP IMPORTANT COLUMNS ONLY
# ====================================

selected_columns = [

    "country",
    "location_name",
    "latitude",
    "longitude",
    "continent",

    "temperature_celsius",
    "humidity",
    "wind_kph",
    "pressure_mb",
    "uv_index",
    "cloud",
    "visibility_km",

    "air_quality_PM2.5",
    "air_quality_PM10",
    "air_quality_Ozone",
    "air_quality_Carbon_Monoxide",

    "condition_text"
]

df = df[selected_columns]

print("\nKolom final:")
print(df.columns.tolist())

# ====================================
# REMOVE DUPLICATES
# ====================================

before_duplicates = df.shape[0]

df = df.drop_duplicates()

after_duplicates = df.shape[0]

print("\nDuplicates removed:")
print(before_duplicates - after_duplicates)

# ====================================
# REMOVE MISSING VALUES
# ====================================

print("\nMissing values sebelum cleaning:")
print(df.isnull().sum())

df = df.dropna()

print("\nMissing values setelah cleaning:")
print(df.isnull().sum())

# ====================================
# REMOVE INVALID VALUES
# ====================================

# remove invalid air quality
df = df[
    (df["air_quality_PM2.5"] >= 0) &
    (df["air_quality_PM10"] >= 0) &
    (df["air_quality_Carbon_Monoxide"] >= 0)
]

# remove unrealistic wind
df = df[df["wind_kph"] < 200]

print("\nDataset setelah remove invalid values:")
print(df.shape)

# ====================================
# RESET INDEX
# ====================================

df = df.reset_index(drop=True)

# ====================================
# FINAL FEATURES FOR ANOMALY
# ====================================

X = df[
    [
        "temperature_celsius",
        "humidity",
        "wind_kph",
        "pressure_mb",
        "uv_index",
        "cloud",
        "visibility_km",
        "air_quality_PM2.5",
        "air_quality_PM10"
    ]
]

print("\nX features:")
print(X.head())

# ====================================
# ANOMALY DETECTION
# ====================================

model = IsolationForest(
    contamination=0.02,
    random_state=42
)

model.fit(X)

predictions = model.predict(X)

df["anomaly"] = predictions

print("\nAnomaly results:")
print(df["anomaly"].value_counts())

# ====================================
# CONTINENT + ANOMALY ANALYSIS
# ====================================

print("\nContinent vs Anomaly:")
print(
    df.groupby(
        ["continent", "anomaly"]
    ).size()
)

# ====================================
# CHECK FINAL DATASET
# ====================================

print("\nFinal dataset shape:")
print(df.shape)

print("\nFinal columns:")
print(df.columns.tolist())

print("\nDuplicate check:")
print(df.duplicated().sum())

print("\nMissing check:")
print(df.isnull().sum())

print("\nDescribe dataset:")
print(df.describe())

print("\nSample anomalies:")
print(
    df[df["anomaly"] == -1]
    .head()
)

# ====================================
# SAVE CLEAN DATASET
# ====================================

df.to_csv(
    "Global_Weather_Clean.csv",
    index=False
)

print("\nGlobal_Weather_Clean.csv berhasil disimpan!")
```

```python
import pycountry_convert as pc

# ====================================
# FIX COUNTRY NAMES
# ====================================

country_fix = {

    "Kyrghyzstan": "Kyrgyzstan",
    "Malásia": "Malaysia",
    "Inde": "India",
    "Mexique": "Mexico",
    "Estonie": "Estonia",
    "Letonia": "Latvia",
    "Komoren": "Comoros",

    "Fiji Islands": "Fiji",
    "Macedonia": "North Macedonia",
    "Cape Verde": "Cabo Verde",

    "USA United States of America":
        "United States",

    "Seychelles Islands":
        "Seychelles",

    "Saint-Vincent-et-les-Grenadines":
        "Saint Vincent and the Grenadines",

    "Democratic Republic of Congo":
        "Congo",

    "Swaziland":
        "Eswatini",

    "Vatican City":
        "Vatican City State",

    "Micronesia":
        "Federated States of Micronesia",

    "Turkey":
        "Türkiye",

    "Russia":
        "Russian Federation",

    "Timor-Leste":
        "East Timor"
}

# ====================================
# APPLY COUNTRY FIX
# ====================================

df_clean["country"] = (
    df_clean["country"]
    .replace(country_fix)
)

# ====================================
# FUNCTION
# ====================================

def get_continent(country_name):

    try:

        country_code = (
            pc.country_name_to_country_alpha2(
                country_name
            )
        )

        continent_code = (
            pc.country_alpha2_to_continent_code(
                country_code
            )
        )

        continent_map = {

            "AF": "Africa",
            "AS": "Asia",
            "EU": "Europe",
            "NA": "North America",
            "SA": "South America",
            "OC": "Oceania"
        }

        return continent_map.get(
            continent_code,
            "Unknown"
        )

    except:
        return "Unknown"

# ====================================
# CREATE CONTINENT COLUMN
# ====================================

df_clean["continent"] = (
    df_clean["country"]
    .apply(get_continent)
)

# ====================================
# CHECK RESULT
# ====================================

print(df_clean["continent"].value_counts())

print("\nFinal Columns:")
print(df_clean.columns.tolist())
```
